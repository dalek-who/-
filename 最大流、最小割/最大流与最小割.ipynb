**割：** s为源点，t为汇点。顶点集V。把V划分为两个集合S，T，其中s∈S，t∈T，这个划分记为CUT(S，T)。称划分CUT(S，T)为一个割  
**割边：** CUT(S,T)是一个割(划分)。若一个边e的一个顶点属于S，另一个顶点属于T，则e称为一个割边。  
**割边的方向：** 从集合S指向集合T的割边称为正向割边，从T指向S的称为逆向割边  
**割的容量：** 割CUT(S,T)上，正向割边的容量之和  
**(有向图的)最小割问题：** 给出图G，源s，目的t，求一个划分CUT(S,T)，即求集合S,T，使得S,T之间的割边的权重（容量）之和最小  
**最小割问题的另一种表述：** 求一个权值（容量）之和最小的边集合E，使得从G中删掉边集合E后，s无法到达t  
  
## 最大流与最小割之间的关系：  
记F为图G上的一个流分布，f为这个流的总流量值。CUT(S,T)是G的一个割(划分),c为这个划分的容量  
**引理1：** 对任意CUT(S,T)，任意流量分布F，总流量值f=S正向割边的容量-S逆向割边的容量  
流量=流出-流入，显然成立  
**推论1：** 任意流F的流量值f，一定小于等于任意割CUT(S,T)的容量c  
f=S正向容量-S逆向容量<=S正向容量=c  
**推论2：** 最大流F的流量f，小于任意割CUT(S,T)的容量c  
  
**引理2：** 假设存在一个割CUT(S,T)，以及一个流量分布F，满足总流量值f=割的容量c，则这个割CUT(S,T)是最小割，这个流分布F是最大流。【注：这样一对割CUT(S,T)和F一定是存在的。后面会给出构造】  
证明：  
最大流：由推论1，对于任意流F\`,其流量f\`<=c=f，因此F是最大流  
最小割：由推论1，对于任意割CUT\`(S\`,T\`),c=f<=c\`,因此CUT(S,T)是最小割  

**定理：最小割CUT的割边：**
* CUT(S,T)的正向割边 <=> 流量=容量的边  
* CUT(S,T)的逆向割边 <=> 流量=0的边  

证明：
* 如果一条正向割边上流量<容量，如果我们把这条边放入割边集合里，则这个割的容量c>流量值f，不满足f=c的条件，得到的不是最小割
* 如果包含流量>0的逆向割边，由推论1的f=S正向容量-S逆向容量<=S正向容量=c，由于逆向容量>0，则f<S正向容量=c，不满足f=c的条件。不是最小割
  
**构造有向图的最小割：**  
1.先用最大流算法求出最大流分布F  
2.先将s加入集合S，之后递归将满足如下条件的节点加入S：
* i∈S,从i到j的 容量\[i\]\[j\]>流量\[i\]\[j\]，将j加入S  
* i∈S,从j到i的 流量\[j\]\[i\]>0，将j加入S  
【**注：**此capacity不是最大流算法中动态变化的残量矩阵，而是初始的容量矩阵】  

3.V-S得到T。于是S,T就是CUT的两个集
