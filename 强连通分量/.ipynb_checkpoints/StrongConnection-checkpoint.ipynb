# 求有向图的强连通分量的Kosaraju算法
- - -
## 定义：
子图G\`是G的一个强连通分量，如果G\`的所有node都可以互相到达  
![示例](./img/强连通分量示意图.PNG)  
如题，这个图有三个强连通分量
## 原理：
**引理1.** 有向图强连通分量，等价于一堆有公共顶点的环路组合在一起(环路上的所有顶点都可绕圈到达，因此环路是强连通的。如上图)  
**引理2.** 有向图G，把所有边的方向逆转得到的逆图rG.rG与G的强连通分量相同(由于强连通分量是由环路组成的，引理2显然成立)  

伪代码：  
```

def strong_connect(G):
    栈stack=[]
    对G进行深度优先搜索,在离开节点v时将v入栈stack #这一步以任何顺序遍历都可以
    
    rstack = stack.reverse()
    rG = G.reverse() #G的逆图
    按rstack的节点顺序，对逆图rG深度优先搜索，每一次无法扩展新节点时，都会得到一个新的强连通分量
    返回得到的所有强连通分量
```  
![原理](./img/原理.PNG)  
算法原理：对于原图G，我们从节点v开始遍历。  
* 它的邻居分三种情况：u1是v的子节点且与v同属于一个强连通分量
* u2是v的子节点且与v同不属于一个强连通分量
* u3是v的父节点(由于v无法到达u3，所有u3一定不属于v的强连通分量)  

后序的深度优先遍历，v先遍历子节点u1，u2，在离开v时才进栈。遍历完G后，栈中的节点顺序：栈底:..u1,..u2,..v,..u3 栈顶  
将stack反转得到rstack: 栈底: ..u3,..v,..u2,..u1 栈顶  
如题可知，如果节点u属于v的强连通分量，则rstack中u一定在v右边(v左边的点一定不在v的强连通分量里)
按rstack的顺序遍历逆图rG:  
* 如图，先遍历u3，u3无法到达v，得到的u3的强连通分量里没有v  
* 如图，遍历v。可以到达u1，v的强连通分量了里有u1。无法到达u2，v的强连通分量里没有u2  

因此，按rstack对逆图rG进行遍历，每次深度优先搜索无法继续扩展时，得到的都是一个强连通分量

## python实现：

````python
import networkx  as nx
import matplotlib.pyplot as plt
import random
````

````python
#随机生成一个图
def random_graph(v_num=10, e_num=20,direction=True):
    G = nx.Graph()
    if direction:
        G=G.to_directed()
    G.add_nodes_from(range(v_num))
    #max_e_num = e_num if e_num<=v_num*v_num else v_num*v_num
    edges = [(random.randint(0,v_num-1),random.randint(0,v_num-1)) for i in range(e_num)]
    G.add_edges_from(edges)
    return G
````

````python
#算法实现：
#因为机考用c，所以这里用类似c风格的python写
#所有append在c中都可以改用一个栈实现

#产生一个图的逆图
def reverse_Graph(G):
    rG = nx.DiGraph()
    rG.add_nodes_from(G.nodes)
    for f,t in G.edges:
        rG.add_edge(t,f)
    return rG
    
#后序的深度优先遍历
def DFS(G,node,stack,visited):
    #print(node)
    if visited[node]:
        return
    visited[node]=True
    #先遍历node的子节点。离开时再把node入栈
    for v in list(G.successors(node)):
        DFS(G,v,stack,visited)
    stack.append(node)
    return

#计算一个图的全部强连通分量
def my_strong_connect(G):
    visited = [False for x in range(G.number_of_nodes())] #标记每个节点是否已经在遍历中被访问了
    stack = [] #后序深度优先遍历时节点加入此栈中
    for v in range(G.number_of_nodes()):
        DFS(G,v,stack,visited)
    
    all_strong_connect = []
    r_G = reverse_Graph(G)
    r_visited = [False for x in range(G.number_of_nodes())]
    rstack = reversed(stack) #反转后的stack
    for v in rstack:
        if r_visited[v]:
            continue
        sc = [] #sc: strong connect，也是一个栈，但是每个sc只收集一个强连通分量的节点
        DFS(r_G,v,sc,r_visited)
        all_strong_connect.append(sc)
    return all_strong_connect
````

````python
#算法正确性测试：
#生成一个随机图
v_num=10
e_num=15
G = random_graph(v_num,e_num)
#结果
print(list(nx.kosaraju_strongly_connected_components(G))) #nx本身封装了这个算法
print(list(my_strong_connect(G))) #我写的kosaraju算法
#绘图
nx.draw_circular(G,with_labels=True)
plt.show()
````
